# Importar bibliotecas

````
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

# Solución a un sistema de tres ecuaciones



```
4x -2y + z = 11
-2x +4y -2z = -16
x -2y + 4z = 17
```

Eliminando ```z``` de las primeras dos ecuaciones ```x = 1```:

y encontramos la linea de intersección de los primeros dos planos (ecuaciones) es igual a ```z = 7 + 2y```



````
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection='3d')

#vamos a graficar a z en función de x y y, para ello creamos primero los planos x, y

x, y = np.linspace(-8, 8,20), np.linspace(-8,8,20)
X, Y = np.meshgrid(x, y)

# despejamos en las tres ecuaciones a Z
Z1 = 11 - 4*X + 2*Y
Z2 = (16 - 2*X + 4*Y) / 2
Z3 = (17 - X + 2*Y) / 4

#superficie de la ecuación 1
ax.plot_surface(X, Y, Z1, alpha=.6, color='cyan',) 

 #superficie de la ecuación 2
ax.plot_surface(X, Y, Z2, alpha=.6, color='yellow')

#linea de intetersección entre la superficies de las primeras dos ecuaciones
ax.plot((1, 1),(-8, 8), (-9,23), lw=3,  c='blue')

#superficie de la ecuación 3
ax.plot_surface(X, Y, Z3, alpha=.3, color='red')

#solución
ax.plot(xs=[1],ys=[-2], zs=[3], c='black',  marker='o')  

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
````
