# Soluci√≥n a un sistema de ecuaciones con dos variables

Supongamos que queremor resolver y luego graficar el siguiente sistema de ecuaciones

`-3x + y = 4`           

` 2x + y = 1`




````
#Matriz de coeficientes
M = np.array([[-3, 1], [2, 1]]) 

# vector con las cantidades al lado derecho de las ecuaciones
b =  np.array([4, 1])

solucion = np.linalg.solve(M, b)
solucion
````

Creamos una funci√≥n 


````
def solucion_ecuaciones_2D(A1, B1, C1, A2, B2, C2):
    '''soluciona un sistema de dos ecuaciones con dos inc√≥nitas. 
    Cada ecuaci√≥n debe estar ordenada de la siguiente forma:
    Ax + By = C'''
    M = np.array([[A1, B1], [A2, B2]])
    b = np.array([C1, C2])
    solucion = np.linalg.solve(M, b)
    return solucion
````

````
def solucion_grafica_2D(A1, B1, C1, A2, B2, C2):
    '''Soluci√≥n gr√°fica a un sistema de dos ecuaciones
    Cada ecuaci√≥n debe estar ordenada de la siguiente forma:
    Ax + By = C'''

    solucion = solucion_ecuaciones_2D(A1, B1, C1, A2, B2, C2)
    lim = 2 * solucion[np.argmax(solucion)]
    x = np.linspace(-lim, lim, 5) 

    #despejamos la y
    y1 = (C1 - A1*x)/B1
    y2 = (C2 - A2*x)/B2

    fig = plt.figure(figsize = (6, 4))

    #crear grafico de linea que representa las ecuaciones
    plt.plot(x, y1, label ='ecuacion 1')     
    plt.plot(x, y2, label = 'ecuacion 2')  
    plt.plot(solucion[0], solucion[1], marker='o', label= 'soluci√≥n')

    #t√≠tulo
    plt.title ('Solucion gr√°fica de un sistema ecuaciones 2 dimensiones')

    #etiquetas de los ejes
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')

    #cuadr√≠cula
    plt.grid(alpha=.8, linestyle='--')

    #agregar etiqueta de valores
    plt.legend( )

    #mostrar la gr√°fica
    plt.show()

    return solucion[0], solucion[1]
````

#  Solucion a un sistemas de ecuaciones con tres variables

Supongamos que queremos resolver y luego graficar el siguientes sistemas de ecuaciones:

```
4x -2y + z =11
-2x +4y -2z = -16
x -2y + 4z = 17
```


````
#Matriz de coeficientes
M = np.array([[4., -2., 1.], [-2., 4., -2.], [1., -2., 4.]]) 

# vector con las cantidades al lado derecho de las ecuaciones
b =  np.array([11, -16, 17])

solucion = np.linalg.solve(M, b)
solucion
````

## Soluci√≥n gr√°fica 

````
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection='3d')

#vamos a graficar a z en funci√≥n de x y y, para ello creamos primero los planos x, y

x, y = np.linspace(-8, 8,50), np.linspace(-8,8,50)
X, Y = np.meshgrid(x, y)

# despejamos en las tres ecuaciones a Z
Z1 = 11 - 4*X + 2*Y
Z2 = (16 - 2*X + 4*Y) / 2
Z3 = (17 - X + 2*Y) / 4

#superficie de la ecuaci√≥n 1
ax.plot_surface(X, Y, Z1, alpha=.6, color='cyan',) 

 #superficie de la ecuaci√≥n 2
ax.plot_surface(X, Y, Z2, alpha=.6, color='yellow')

#linea de intetersecci√≥n entre la superficies de las primeras dos ecuaciones
ax.plot((1, 1),(-8, 8), (-9,23), lw=3,  c='blue')

#superficie de la ecuaci√≥n 3
ax.plot_surface(X, Y, Z3, alpha=.3, color='red')

#soluci√≥n
ax.plot(xs=[1],ys=[-2], zs=[3], c='black',  marker='o')  

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
````

#  Solucion a un sistemas de ecuaciones en m√°s de 3 dimensiones

Considere el siguiene sistema de ecuaciones:

```
a + b + c + d = 4
2a + 3b -c + d = 10
a - b + c - d = -2
a + 2b + 3c - d = -1
```
Vamos a encontrar el valor de a, b, c y d usando la funci√≥n `solve` de numpy


````
coeficientes = np.array([
                         [1,  1,  1,  1],
                         [2,  3, -1,  1],
                         [1, -1,  1, -1],
                         [1,  2,  3, -1]
                         ])

resultados = np.array([4, 10, -2, -1])

a, b, c, d = np.linalg.solve(coeficientes, resultados)

print('a = {:.1f} \nb = {:.1f} \nc = {:.1f} \nd = {:.1f}'.format(a, b, c, d))
````
